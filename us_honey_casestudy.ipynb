Q2.which are the top 5 honey producing states in us ?

top and bottom 5 states in honey production

```python
x=df.groupby('state')['production'].sum().reset_index()
x

#['production'].sum()-Aggregates the 'production' column by summing up the values for each state.
#.reset_index()-Converts the grouped result back into a DataFrame with a proper index.
```

```python
top5=x.sort_values(by='production',ascending=False).head(5)
top5
```

```python
bottom5=x.sort_values(by='production',ascending=False).tail(5)
bottom5
```

Q3.what is the change in mean average price of honey from 1995 to 2021?

```python
Y=df.groupby('year')['average_price'].mean()
Y
```

```python
change=Y.loc[1995]-Y.loc[2021]
change
```

Q4.which was the year when the production of honey in the whole us was the highest?

```python
a=df.groupby('year')['production'].sum().reset_index()
a
```

```python
a.sort_values(by='production',ascending=False).head(1)
```

Q5.from the above inference we get the production was highest in the year 2000 now infer which state was having contribution in that year?

```python
d = df.groupby(['state','year'])['production'].sum().reset_index()
d
```

```python
e =d.loc[d['year']==2000]
e
```

```python
e.sort_values(by='production',ascending=False).head(1)
```
